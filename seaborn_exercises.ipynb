4. Load the chipotle dataset from SQL, create a bar chart that shows the 4 most popular items and the revenue produced by each.

    A: Not conclusive but lower levels of education resulted in higher fertility numbers. More education would lead to less fertility.

```python
url_chipotle = env.get_db_url(env.user,env.password,env.host,'chipotle')
query_chipotle = 'select * from orders'
chipotle = pd.read_sql(query_chipotle,url_chipotle)
chipotle
```

```python
chipotle['item_price'] = chipotle.item_price.str.strip('$').astype(float)
```

```python
top_4 = chipotle.groupby('item_name').quantity.sum().nlargest(4)
```

```python
all_prices = chipotle.groupby('item_name').item_price.sum()
```

```python
merged_df = pd.merge(top_4, all_prices, on='item_name',how='inner')
merged_df= merged_df.reset_index()
merged_df
```

```python
sns.barplot(data=merged_df, x='item_price', y='item_name')
plt.title('Top 4')
plt.xlabel('Prices')
plt.ylabel('Name of Item')
```

```python
chipotle.groupby('item_name').total_price.sum().sort_values(ascending=False).head(4).plot(kind='barh').set(title='chipotle',
                                                                                                            xlabel='sales',
                                                                                                            ylabel='item name')
```

Another attempt using melt to melt subjects together

```python
subject_group = sleepstudy.reset_index().melt(id_vars=['index','Subject'], var_name='col')
# melting subject groups to be together
subject_group['Subject'] = subject_group.Subject.astype(str)
```

```python
o = sns.lineplot(x='index',y='value',data=subject_group, hue='Subject',estimator='mean')
plt.xlabel('Subjects')
plt.ylabel('Reactions')
plt.figure(figsize=(2,5))
sns.move_legend(o, 'upper left',bbox_to_anchor=(1,1))
```
