```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
print(tips.info())
tips.isnull().sum()
```

```python
sns.relplot(x='total_bill',y='tip', data=tips, marker = '+', s=100)
plt.title('total bill vs tips')
```

```python
sns.relplot(x='total_bill',y='tip',data=tips, col='size',col_wrap=3)
```

```python
sns.relplot(x='total_bill',y='tip',data=tips, hue='size',style='size')
```

```python
sns.histplot(tips['total_bill'],kde=True)
```

```python
sns.boxplot(y='tip',data=tips, x='size')
plt.title('Tips')
```

```python
plt.figure(figsize=(12,10))
plt.suptitle('boxplots w sns')

plt.subplot(221)
sns.boxplot(data=tips,y='tip')

plt.subplot(222)
sns.boxplot(data=tips, x = 'tip')
plt.title('rotated')

plt.subplot(223)
sns.boxplot(data=tips, y='tip', x='size')
plt.title('size to tip')

plt.subplot(224)
sns.boxplot(data=tips, y='time',x='tip',hue='smoker')
plt.title('smoker vs non smoker')
```

```python
sns.pairplot(tips)
```

```python
data = pd.crosstab(tips['time'], tips['size'])
sns.heatmap(data, cmap=plt.cm.Greens)
```

```python
sns.jointplot(data=tips, x='total_bill', y= 'tip')
```
