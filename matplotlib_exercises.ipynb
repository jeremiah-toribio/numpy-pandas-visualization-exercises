```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import math
```

```python
x = np.linspace(-2,2,100)
y = x**2 - x + 2
```

```python
plt.plot(x,y)
plt.annotate('Origin', xy=(0,0), xytext=(0,2), arrowprops={'arrowstyle':'fancy','facecolor': 'blue'})
plt.show()
```

```python
y = np.sqrt(x)
plt.scatter(x,y)
```

```python
y = x ** 3
plt.scatter(x,y)
```

```python
x = np.linspace(0,2,100)
y = 2 ** x
plt.plot(x,y)
```

```python
x = np.linspace(0,2,100)
y = 1/(x+1)
plt.plot(x,y)
```

```python
x = np.linspace(0,2,100)
y = np.sqrt(x)
plt.subplot(221)
plt.scatter(x,y,s=10,c='red')

x = np.linspace(0,2,100)
y = x ** 3
plt.subplot(222)
plt.scatter(x,y)

x = np.linspace(0,2,100)
y = 2 ** x
plt.subplot(223)
plt.scatter(x,y)

x = np.linspace(0,2,100)
y = 1/(x+1)
plt.subplot(224)
plt.scatter(x,y)
```

```python
x = np.linspace(0,2,100)
y = np.sqrt(x)
plt.scatter(x,y,s=10,c='red', label = '$\sqrt{x}$')

a = np.linspace(0,2,100)
b = a ** 3
plt.scatter(a,b,s=10,c='blue', label = 'b = $a^3$')

c = np.linspace(0,2,100)
d = 2 ** c
plt.scatter(c,d,s=10,c='green', label = 'd = $2^c$')

j = np.linspace(0,2,100)
k = 1/(x+1)
plt.scatter(j,k,s=10,c='yellow', label = 'k = 1/x+1')

plt.legend(loc='upper left')

plt.title('Four Algos')
plt.show()

```
