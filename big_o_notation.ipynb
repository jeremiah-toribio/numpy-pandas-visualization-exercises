```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import math
```

```python
x = range(1,3,1)   # np.linspace(1,2,100)
y = [(0*x) + 1 for x in x] # math.log(x)

plt.plot(y,label='O(1)')

a = [math.log(x) for x in x]
plt.plot(a,label='O(log x)')

b = x
plt.plot(b, label = 'O(x)')

c = np.log(x) * x
plt.plot(c, label = 'O(x log x)')
#d = [(math.log(x) * x) for x in x]
#plt.plot(d)

d = [x ** 2 for x in x]
plt.plot(d, label='O($x^2$)')

e = [2 ** x for x in x]
plt.plot(e, label='O($2^x$)')

f = [math.factorial(x) for x in x]
plt.plot(f,label='O(n!)')

g = [x ** x for x in x]
plt.plot(g, label='O($x^x$)')

plt.legend(loc='upper left')
plt.title('Big O Notation')
plt.xlabel('Elements')
plt.ylabel('Operations')
```
